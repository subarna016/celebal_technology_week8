```python
# check for missing values
data.isnull().sum()
```

```python
# replace missing values with the mode
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Married'].fillna(data['Married'].mode()[0], inplace=True)
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)
```

```python
data['Loan_Amount_Term'].value_counts()
```

```python
# replace missing value with the mode
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0], inplace=True)
```

```python
# replace missing values with the median value due to outliers
data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# before log transformation
ax1 = plt.subplot(121)
data['LoanAmount'].hist(bins=20, figsize=(12,4))
ax1.set_title("Data")
```

```python
# Removing skewness in LoanAmount variable by log transformation
data['LoanAmount_log'] = np.log(data['LoanAmount'])
```

```python
# after log transformation
ax1 = plt.subplot(121)
data['LoanAmount_log'].hist(bins=20, figsize=(12,4))
ax1.set_title("Data")
```

```python
# combine Applicant Income and Coapplicant Income into a new variable
data['Total_Income'] = data['ApplicantIncome'] + data['CoapplicantIncome']
```

```python
# check the distribution of Total Income
# before log transformation

fig = plt.figure(figsize=(14, 4))
ax1 = plt.subplot(121)
sns.distplot(data['Total_Income'])
 
ax1.set_title("Data")
```

```python
# log transformation
data['Total_Income_log'] = np.log(data['Total_Income'])
```

```python
# after log transformation
fig = plt.figure(figsize=(14, 4))
ax1 = plt.subplot(121)
sns.distplot(data['Total_Income_log'])
ax1.set_title("Data")
```

```python
# create EMI feature
data['EMI'] = data['LoanAmount'] / data['Loan_Amount_Term']
```

```python
# check the distribution of EMI
fig = plt.figure(figsize=(14, 4))
ax1 = plt.subplot(121)
sns.distplot(data['EMI'])
ax1.set_title("Data")
```

```python
# create new "Balance Income" variable
data['Balance Income'] = data['Total_Income'] - (data['EMI']*1000)
```

```python
# check the distribution of EMI
# before log transformation
fig = plt.figure(figsize=(14, 4))
ax1 = plt.subplot(121)
sns.distplot(data['Balance Income'])
ax1.set_title("Data")
```
